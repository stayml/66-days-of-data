```python
# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np 

%matplotlib inline
```

```python
# sample data
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
# creating a plot
fig, ax = plt.subplots(
    
    # kwargs typed manually for learning
    
    nrows=2,
    ncols=2,
    figsize=(10,5)
    
)

# setting titles for axes

ax[0, 0].set_title("I'm ax[0, 0]")
ax[0, 1].set_title("I'm ax[0, 1]")

ax[1, 0].set_title("I'm ax[1, 0]")
ax[1, 1].set_title("I'm ax[1, 1]")

# set xmin, xmax for x axis of ax[0, 1] - useful when data not starting at 0

ax[0, 1].set(xlim=(-50, 50), ylim=(20, 40))

# set margins

ax[0, 0].margins(.4, .4)

# scatter some data

ax[0, 0].scatter(x**2, y**4)
ax[1, 1].scatter(x, y)

fig.suptitle('Title for')

fig.tight_layout()
```
